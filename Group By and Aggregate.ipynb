**Group By and Aggregate**

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\gerar\Downloads\Flavors.csv")
df
```

```python
base_flavor = df.groupby('Base Flavor')
# Seleccionar solo columnas numéricas
numerical_columns = df.select_dtypes(include='number').columns

# Calcular la media solo para las columnas numéricas
base_flavor_mean = base_flavor[numerical_columns].mean()
```

```python
base_flavor_mean
```

```python
base_flavor_count = base_flavor[numerical_columns].count()
base_flavor_count
```

```python
base_flavor.min()
```

```python
base_flavor.max()
```

```python
base_flavor.sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'count', 'min', 'max']})
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'count', 'min', 'max'], 'Texture Rating': ['mean', 'count', 'min', 'max']})
```

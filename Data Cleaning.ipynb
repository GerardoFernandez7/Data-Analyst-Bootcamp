**Data Cleaning**

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\gerar\Downloads\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('./_')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]', '', regex=True) # Quitar signos en medio
df 
```

```python
df['Phone_Number'].apply(lambda x: str(x)) # Convertir la col en str
# Aplicar el formato adecuado a los nmeros de telfono, manejando los NaN
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] if isinstance(x, str) and len(x) == 10 else x)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].fillna('')
df
```

```python
df['Address'].str.split(',', n = 2, expand = True)
```

```python
df[['Street_Adress', 'State', 'Zip_Code']] = df['Address'].str.split(',', n = 2, expand = True)
df
```

```python
df = df.drop(columns = 'Address')
df
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y', regex=True)
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N', regex=True)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y', regex=True)
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N', regex=True)
df
```

```python
df = df.fillna('')
df
```

```python
df = df.replace('N/a', '')
df
```

**Ahora no queremos que el call center tenga los contactos de las personas que no quieren ser contactadas**

```python
df.drop(df[df['Do_Not_Contact'] == 'Y'].index, inplace=True)
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.strip()
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('', 'N')
df
```

```python
df = df.reset_index()
df
```

```python
df = df.drop(columns = 'index')
df
```

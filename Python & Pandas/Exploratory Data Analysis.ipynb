```python
# In which way are others continents growing Faster
# Filtrar solo las columnas numéricas
numeric_df = df.select_dtypes(include='number')

# Añadir la columna 'Continent' al DataFrame numérico
numeric_df['Continent'] = df['Continent']

# Calcular la media agrupada por continente y ordenar por '2022 Population'
df2 = numeric_df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population", ascending = False)

df2
```

 **Merge, Join & Concatenate**
 ![image.png]

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\gerar\Downloads\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\gerar\Downloads\LOTR 2.csv")
df2
```

```python
df1.merge(df2) #Inner Join
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner', on = ['FirstName', 'FellowshipID'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

**Joins se usa mas que nada cuando se trabaja con index**

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df4
```

**Concatenate es basicamente poner uno encima de otro en vez de ponerlo uno al lado de otro como los anteriores**

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join = 'inner')
```

```python
pd.concat([df1, df2], join = 'outer')
```

```python
pd.concat([df1, df2], join = 'outer', axis = 1)
```
